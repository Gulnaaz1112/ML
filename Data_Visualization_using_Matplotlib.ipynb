```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([12,13,15,18,19,23])   # x-coordinates value
y = np.array([56,78,28,89,76,56])   # y-coordinates value
z = np.array([67,34,89,54,98,23])
ax = plt.axes()
ax.set_facecolor('#c3f7f6')
plt.plot(x,y,c ='brown',marker='*',label='Team 1')
plt.plot(x,z,c ='#27585e',marker='+',label='Team 2')
plt.legend()
plt.xlim(0)
plt.xlabel('X-axis values')
plt.ylabel('Y-axis values')
plt.title('My Line Graph')
plt.show()
```

```python
# Scatter Plot

import matplotlib.pyplot as plt
import numpy as np
x = np.array([12,13,15,18,19,23])   # x-coordinates value
y = np.array([56,78,28,89,76,56])   # y-coordinates value
z = np.array([67,34,89,54,98,23])
plt.scatter(x,y,s=200,alpha =0.5)
plt.scatter(x,z)
plt.show()
```

```python
people = ['Mr X','Mr Y','Mr Z','Mr A']
weight = [45,56,47,57]
plt.bar(people,weight,color=['g','b','r','k'])
plt.xlabel('People Name')
plt.ylabel('Weight of people')
plt.title('Bar Graph')
plt.show()
```

```python
plt.figure(figsize=(15,10))
x = np.array([12,13,15,18,19,23])   # x-coordinates value
y = np.array([56,78,28,89,76,56])   # y-coordinates value
z = np.array([67,34,89,54,98,23])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Sub Graph 1")
plt.subplot(1,2,2)
plt.plot(x,z)
plt.title("Sub Graph 2")

plt.suptitle("Main Graph")
plt.show()
```
